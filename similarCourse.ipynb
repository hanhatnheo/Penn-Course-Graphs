```python
import json
```

```python
with open("data.json", "r") as read_file:
    data = json.load(read_file)
```

```python
keys = data.keys()
```

```python
keys = list(keys)
keys.reverse()
```

```python
codeDescription = {}
```

```python
for key in keys:
  codeDescription[key] = data[key]['description']
```

```python
texts = codeDescription.values()
texts = [jieba.lcut(text) for text in texts]
dictionary = corpora.Dictionary(texts)
feature_cnt = len(dictionary.token2id)
corpus = [dictionary.doc2bow(text) for text in texts]
tfidf = models.TfidfModel(corpus) 
index = similarities.SparseMatrixSimilarity(tfidf[corpus], num_features = feature_cnt)

for key in keys:
    print("Considering key " + key)
    similar = []
    keyword = codeDescription[key]
    kw_vector = dictionary.doc2bow(jieba.lcut(keyword))
    sim = index[tfidf[kw_vector]]
    
    similar = [list(codeDescription.keys())[i] for i, s in enumerate(sim) if s >= 0.3 and list(codeDescription.keys())[i] != key]
    
    data[key]['similar'] = similar
    print(similar)       

json_object = json.dumps(data)

with open("data_v2-Copy1.json", "w") as outfile:
    outfile.write(json_object)
            
```
